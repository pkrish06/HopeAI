```python
# discount 10 percent exceed 100
total = 0
items = int(input("enter the number of item"))
for i in range(0, items):
    amount = int(input("enter the item" + str(i+1) + " amount:"))
    total = total + amount

if (total >= 100):
    discount = (total*.1)
    finalAmount = (total - discount)
else:
    discount = 0
    finalAmount = total

print("Your total amount " + str(total))
print("Your final bill amount is " + str(finalAmount) + " and discount is " + str(discount))
```

```python
#bank withdrawal
minBalance = 1000
currentBalence = int(input("Enter the current balance:"))
withDrawalAmount = int(input("Enter the Amout:"))
if (currentBalence-withDrawalAmount >= minBalance):
    print("Withdrawal allowed")
else:
    print("No minimum balance")
```

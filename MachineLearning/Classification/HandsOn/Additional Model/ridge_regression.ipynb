```python
#importing the Libraies
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Reading the Dataset
dataset = pd.read_csv('insurance_pre.csv')
dataset
```

```python
dataset.columns
```

```python
dataset=pd.get_dummies(dataset, dtype=int, drop_first=True)
dataset
```

dataset

```python
indep=dataset[['age', 'sex_male', 'bmi', 'children', 'smoker_yes']]
dep=dataset[['charges']]
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import Ridge

param_grid = {'solver':['auto', 'svd', 'cholesky', 'lsqr', 'sparse_cg', 'sag', 'saga', 'lbfgs'],
'max_iter':[10,100,1000]}

grid = GridSearchCV(Ridge(), param_grid, refit = True, verbose = 3,n_jobs=-1)
# fitting the model for grid search
grid.fit(indep, dep)
```

```python
re=grid.cv_results_
print("The R_score value for best parameter {}:".format(grid.best_params_))
print("The R_score value {}:".format(grid.best_score_))
```

```python
table=pd.DataFrame.from_dict(re)
```

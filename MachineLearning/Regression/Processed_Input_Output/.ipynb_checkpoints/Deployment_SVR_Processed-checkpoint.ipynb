```python
import pickle

# Load scaler
with open("standard_scaler_SVR.sav", "rb") as file:
    scaler = pickle.load(file)

# Load model
with open("SVR_Processed_IO_Deployment_Model.sav", "rb") as file:
    model = pickle.load(file)

# User input
age = int(input("Age: "))
sex_male = int(input("Sex Male (0 or 1): "))
bmi = float(input("BMI: "))
children = int(input("Children: "))
smoker_yes = int(input("Smoker (0 or 1): "))

# Scale input
processed_input = scaler.transform([[age, sex_male, bmi, children, smoker_yes]])

# Predict
result = model.predict(processed_input)

print("Predicted Insurance Charge:", result[0])
```

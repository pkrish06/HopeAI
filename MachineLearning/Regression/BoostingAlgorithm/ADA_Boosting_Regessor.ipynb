```python
#importing the Libraies
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Reading the Dataset
dataset = pd.read_csv('insurance_pre.csv')
dataset
```

```python
dataset.columns
```

```python
dataset=pd.get_dummies(dataset, dtype=int, drop_first=True)
dataset
```

dataset

```python
indep=dataset[['age', 'sex_male', 'bmi', 'children', 'smoker_yes']]
dep=dataset[['charges']]
```

```python
indep
```

```python
dep
```

```python
#split into training set and test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(indep, dep, test_size = 1/3, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
X_train
```

```python
from sklearn.ensemble import AdaBoostRegressor
regressor = AdaBoostRegressor(random_state=0, n_estimators=100)
regressor.fit(X_train, y_train)#y=W*x1+b0 for this equation we got value for b1 and bo
```

```python
# Viewing the b1 and bo value
estimators=regressor.estimators_
print("estimators={}".format(estimators))
estimator_weights_=regressor.estimator_weights_
print("estimator_weights_={}".format(estimator_weights_))
```

```python
y_pred=regressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
```
